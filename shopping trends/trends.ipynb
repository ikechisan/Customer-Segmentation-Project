## Numerical Overview
* **Age:** The average age is 44 with standard deviation of 15.2 indicating a diverse age range between 18 and 70. The median age is also 44, suggesting that the mean is not significantly affected by outliers.
* **Purchase Amount:** The average purchase amount is approximately $60 with a standard deviation of 23.68.
* **Review Rating:** The reviews range from 2.5 to 5 which suggests generally positive reviews with an average of 3.75 indicating that must customers are satisfied with their purchases.
* **Previous Purchases:** The average number of previous purchases is 25.35 with a minimum of 1 purchase and a standard deviation of 14.45. 


```python
gender_counts = shopping_trends['Gender'].value_counts()
gender_counts
```

```python
frequency_counts = shopping_trends['Frequency of Purchases'].value_counts()
frequency_counts
```

```python
# Combine "Bi-Weekly" and "Fortnightly" into "Every Two Weeks"
shopping_trends['Frequency of Purchases'] = shopping_trends['Frequency of Purchases'].replace(['Bi-Weekly', 'Fortnightly'], 'Every 2 Weeks')

# Combine "Every 3 Months" and "Quarterly" into "Quarterly"
shopping_trends['Frequency of Purchases'] = shopping_trends['Frequency of Purchases'].replace(['Every 3 Months', 'Quarterly'], 'Every 3 Months')
```

```python
preferred_payment_counts = shopping_trends['Preferred Payment Method'].value_counts()
preferred_payment_counts
```

```python
subscription_counts = shopping_trends['Subscription Status'].value_counts()
subscription_counts
```

```python
category_counts = shopping_trends['Category'].value_counts()
category_counts
```

```python
combined_counts1 = pd.concat([gender_counts, frequency_counts, preferred_payment_counts, subscription_counts, category_counts], axis=0)
combined_counts1
```

```python
shopping_trends['Customer Segment'] = 'Other'
shopping_trends.loc[(shopping_trends['Frequency of Purchases'].isin(['Weekly', 'Every 2 Weeks'])) & (shopping_trends['Subscription Status'] == 'Yes') & (shopping_trends['Purchase Amount (USD)'] > 75), 'Customer Segment'] = 'VIP Customers'
shopping_trends.loc[(shopping_trends['Frequency of Purchases'].isin(['Weekly', 'Every 2 Weeks', 'Monthly'])) & (shopping_trends['Subscription Status'] == 'Yes') & (shopping_trends['Purchase Amount (USD)'] >= 50) & (shopping_trends['Purchase Amount (USD)'] < 75), 'Customer Segment'] = 'Loyal Customers'
shopping_trends.loc[(shopping_trends['Frequency of Purchases'].isin(['Every 3 Months', 'Annually'])) & (shopping_trends['Subscription Status'] == 'Yes'), 'Customer Segment'] = 'Inactive Subscribers'
shopping_trends.loc[(shopping_trends['Frequency of Purchases'].isin(['Every 3 Months', 'Monthly'])) & (shopping_trends['Subscription Status'] == 'No') & (shopping_trends['Purchase Amount (USD)'] >= 60) & (shopping_trends['Purchase Amount (USD)'] < 80), 'Customer Segment'] = 'Regular Customers'
shopping_trends.loc[(shopping_trends['Frequency of Purchases'].isin(['Weekly', 'Every 2 Weeks', 'Monthly'])) & (shopping_trends['Subscription Status'] == 'No') & (shopping_trends['Purchase Amount (USD)'] >= 60) & (shopping_trends['Purchase Amount (USD)'] < 80), 'Customer Segment'] = 'Potential High-Value Customers'
shopping_trends.loc[(shopping_trends['Frequency of Purchases'].isin(['Weekly', 'Every 2 Weeks', 'Monthly'])) & (shopping_trends['Subscription Status'] == 'No') & (shopping_trends['Purchase Amount (USD)'] > 80), 'Customer Segment'] = 'High-Value Non-Subscribers'
shopping_trends.loc[(shopping_trends['Frequency of Purchases'].isin(['Weekly', 'Every 2 Weeks', 'Monthly'])) & (shopping_trends['Subscription Status'] == 'No') & (shopping_trends['Purchase Amount (USD)'] < 40) & (shopping_trends['Review Rating'] > 3), 'Customer Segment'] = 'Low-spending Frequent Shoppers'
shopping_trends.loc[(shopping_trends['Frequency of Purchases'].isin(['Annually', 'Every 3 Months'])) & (shopping_trends['Subscription Status'] == 'No') & (shopping_trends['Purchase Amount (USD)'] < 40) & (shopping_trends['Review Rating'] < 3.75), 'Customer Segment'] = 'At-Risk Customers'
shopping_trends
```
